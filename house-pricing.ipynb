```python
def print_large_missingno(data, amount=5):
    chunk = []
    columns = []
    n = 0
    total = 0
    for col in data.columns:
        columns.append(col)
        n = n + 1
        total = total + 1
        if n % amount == 0 or total == len(data.columns):
            chunk.append(columns)
            columns = []
            n = 0
    
    [msno.matrix(data[cols]) for cols in chunk]
```

```python
print_large_missingno(data, amount=12)
```

```python
bad_columns = list()

rows = data.shape[0]
for col in data.columns:
    nulls = data[col].isnull().sum()
    if nulls > 0 and rows / nulls < 0.5:
        bad_columns.append(col)
```
