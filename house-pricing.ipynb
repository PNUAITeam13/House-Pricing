```python
def print_large_missingno(data, amount=5):
    chunk = []
    columns = []
    n = 0
    total = 0
    for col in data.columns:
        columns.append(col)
        n = n + 1
        total = total + 1
        if n % amount == 0 or total == len(data.columns):
            chunk.append(columns)
            columns = []
            n = 0
    
    [msno.matrix(data[cols]) for cols in chunk]
```

```python
from data_transformer.DataTransformer import DataTransformer

def transform_ordinal(data, transformer=DataTransformer()):
    s = (data.dtypes == 'object')
    object_columns = list(s[s].index)
    
    return transformer.use_ordinal_encoder(data, object_columns).get_data()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(data_featured, y, train_size=0.8)
```

```python
from xgboost import XGBRegressor, XGBClassifier
from sklearn.metrics import mean_absolute_error, accuracy_score, mean_squared_error, f1_score

def run_xgboost_regressor(X_train, y_train, X_valid, y_valid, estimators=[1000]):
    best_model = None
    best_score = None
    for n_estimator in estimators:
        xgboost_regressor = XGBRegressor(n_estimators=n_estimator, learning_rate=0.06, max_depth=10)
        xgboost_regressor.fit(X_train, y_train)
        predictions = xgboost_regressor.predict(X_valid)
        mae = mean_absolute_error(y_valid, predictions)
        print(f"XGBOOST Regressor ({n_estimator}): MAE - {mae}")

        if best_model is None or mae < best_score:
            best_model = xgboost_regressor
            best_score = mae

    return best_model

def run_xgboost_classifier(X_train, y_train, X_valid, y_valid, estimators=[1000]):
    best_model = None
    best_score = None
    for n_estimator in estimators:
        xgboost_classifier = XGBClassifier(objective='multi:softmax', num_class=len(set(y_train)), n_estimators=n_estimator, 
                                           learning_rate=0.1, max_depth=5, verbosity=2)
        xgboost_classifier.fit(X_train, y_train)
        predictions = xgboost_classifier.predict(X_valid)
        a_score = accuracy_score(y_valid, predictions)
        print(f"XGBOOST Classifier: Accuracy Score - {a_score}, F1 Score - {f1_score(y_valid, predictions)}")
        if best_model is None or a_score > best_score:
            best_model = xgboost_classifier
            best_score = a_score

    return best_model
```

```python
xgboost_regressor = run_xgboost_regressor(X_train, y_train, X_valid, y_valid, estimators=[500, 600, 700, 800, 1000])
```
